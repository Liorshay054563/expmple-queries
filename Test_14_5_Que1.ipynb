```python

import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# Question 1 - part a- info, describe, head
customers_df = pd.read_csv('Customers.csv')
customers_df.info()
customers_df.describe().T
customers_df.head()

orders_df = pd.read_csv('orders.csv')
orders_df.info()
orders_df.describe().T
orders_df.head()

order_item_df = pd.read_csv('order_item.csv')
order_item_df.info()
order_item_df.describe().T
order_item_df.head()

items_df = pd.read_csv('Items.csv')
items_df.info()
items_df.describe().T
items_df.head()
```

```python
# Question 1 - duplicates handling .

dup_counts = items_df.groupby('item_name')['id'].count()
display(dup_counts)
display(dup_counts[dup_counts > 1])

# duplications
dup = items_df[items_df['item_name'].duplicated(keep=False)].copy()

## new column count not-null
dup['non_nan_count'] = dup.notna().sum(axis=1)

## sort from big - small
dup.sort_values(['non_nan_count','item_name'], ascending=[False, True], inplace=True)
print("duplicates===")
display(dup)

## drop duplicates and keep the first
kept = dup.drop_duplicates(subset=['item_name'], keep='first')

#saving the removed
removed = dup[~dup['id'].isin(kept['id'])]
print("remove ===")
display(removed)

## 4 remove dup from items
after_remove = items_df[~items_df['id'].isin(removed['id'])]
after_remove

## check if clean from dup
dup_counts = after_remove.groupby('item_name')['id'].count()
display(dup_counts[dup_counts > 1]) # if true = no dup

## remove from items_df
# items = items_df.copy() # if i dont want too change the original df
items_df = items_df[~items_df['id'].isin(removed['id'])] 
```

```python
order_item_df = pd.read_csv('order_item.csv')
# merged between the removed and the kept 
merged = removed.merge(kept, how='inner', on='item_name', suffixes=('_removed', '_kept'))
display(merged)
# connect the id's that duplicates
result_map = dict(zip(merged['id_removed'], merged['id_kept']))
display(result_map)
print("check who many dup id's ")
display(order_item_df['id'].isin(removed['id']).sum())

## update -replace the order_items id's by the dict
order_item_df['id'] = order_item_df['id'].replace(result_map)

# # check how many remove still exist
order_item_df['id'].isin(removed['id']).sum()  # if zero = all id that removed replace succsesfuly
```
